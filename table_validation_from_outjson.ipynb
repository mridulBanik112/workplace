```python
import json
import cv2
import os
from PIL import Image, ImageDraw, ImageOps
import numpy as np


def drawn_image(image,cord_list):
    
    for i, box in enumerate(cord_list):
        x1, y1, x2, y2, x3, y3, x4, y4= box
        x_min = min(x1, x2, x3, x4)
        x_max = max(x1, x2, x3, x4)
        y_min = min(y1, y2, y3, y4)
        y_max = max(y1, y2, y3, y4)
        cv2.rectangle(image,(x_min,y_min),(x_max,y_max),(0,0,0),4)
    return image

def plot_boxes(img, boxes):
    '''plot boxes on image
       input: 
            img: PIL image
            boxes: (n,8), list of boxes. each box has 8 values
       output:
            img: Drawn image
    '''
    if boxes is None:
        return img
    
    draw = ImageDraw.Draw(img)
    for box in boxes:
        draw.line([box[0], box[1], box[2], box[3], box[4], box[5], box[6], box[7], box[0], box[1]], fill=(0,0,0), width=4)
    return img


def json_data_rectangle(image,data):
    cord_list=[]
    
    for ids in data["segments"]:
        if(data['segments'][ids]['type'] == 'paragraph' or data['segments'][ids]['type'] == 'marginalia' or data['segments'][ids]['type'] == 'TableRegion'):
                all_cord = data['segments'][ids]["textlines"]
                #print(all_cord)
                l = []
                
                for points in all_cord:
                    l.append(all_cord[points]['points'])
                    #print(l)
                for each_cords in l:
                    x_y_cord = []
                    for cord in each_cords:
                        
                        x_y_cord.append(int(cord['x']))
                        x_y_cord.append(int(cord['y']))
                        #print(x_y_cord)
                    cord_list.append(x_y_cord)
                print(cord_list)
                print(len(cord_list))
    #img = Image.open(img_file).convert('RGB')
    #draw = ImageDraw.Draw(img)
    #for box in cord_list:
        #draw.line([box[0], box[1], box[2], box[3], box[4], box[5], box[6], box[7], box[0], box[1]], fill=(0,0,0), width=4)
    #img2 = np.array(img)
    image=drawn_image(image,cord_list)
    #cv2.namedWindow("output", cv2.WINDOW_NORMAL)        # Create window with freedom of dimensions
    #im = cv2.imread("earth.jpg")                        # Read image
    #imS = cv2.resize(img2, (1000, 1200))               # Resize image
    #cv2.imshow("output", imS)                            # Show image
    #cv2.waitKey(0)     
    return image
    



##############FOR TEST#################

# Opening JSON file
json_file_path =r"E:\ADRIL[Office]\Code for easeness\PipelineTask\return_json71.json" 
img_file=r"E:\ADRIL[Office]\Code for easeness\PipelineTask\Test\71.png"
outimg_path=r"E:\ADRIL[Office]\Code for easeness\PipelineTask\Test\outdemo"
out_img = os.path.join(outimg_path, os.path.basename(img_file))
image = cv2.imread(img_file)
with open(json_file_path, encoding="utf8") as json_file:
    data = json.load(json_file)
    image=json_data_rectangle(image,data)

#plotted_image= drawn_image(image,cord_list)
cv2.imwrite(out_img,image)
```
